```python
import pandas as pd 
import numpy as np
import seaborn as sns 
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("advertising.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
X = df[["TV","Radio","Newspaper"]]
y = df[["Sales"]]
X
y
```

```python
df.corr()
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=0, test_size=60)
```

```python
X_train
```

```python
model = LinearRegression()
```

```python
# supervised learning 

model.fit(X_train,y_train)
```

```python
y_pred_test = model.predict(X_test)
```

```python
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
r2_score(y_test,y_pred_test)
```

```python
# Here we are making function and using our model the one we train above 

def sales_prediction():
    # here we take budget as input from user (i.e Tv,radio,newspaper)
    tv = float(input("Enter TV ad budget: "))
    radio = float(input("Enter Radio ad budget: "))
    newspaper = float(input("Enter Newspaper ad budget: "))
    # we check min & max of our data and then i decide to write this line cuz of we enter input(budget) high model might not know about that!
    # so the model become unreliable
    if tv > 296.4 or radio > 49.6 or newspaper > 114.0:
        print("Warning: Your input is outside the training data range. Prediction might be unreliable!")

    new_data = np.array([[tv,radio,newspaper]])
    prediction = model.predict(new_data)

    print(f"The predicted sales are: ${prediction[0][0]:.2f}")

    


```
